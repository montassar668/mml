```
#Convertire le colonne bank account au format entier (donn√©es num√©riques) en utilisant label encoder from sklearn
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
df_Train['bank_account'] = lc.fit_transform(df_Train['bank_account'])
df_Train
```

```
#Ajouter une autre colonne nom√©e year pour les deux dataframe
df_Train['year_'] = df_Train['year']
df_test['year_'] = df_test['year']

# Convertit les colonnes household_size,age_of_respondent,year en flottant
float_array = df_Train[['household_size', 'age_of_respondent', 'year_']].values.astype(float)
float_array = df_test[['household_size', 'age_of_respondent', 'year_']].values.astype(float)
```

```
# convertir des entit√©s cat√©gorielles en entit√©s num√©riques
df_Train['country_'] = df_Train['country']
df_test['country_'] = df_test['country']

categ = ['relationship_with_head', 'marital_status', 'education_level', 'job_type', 'country_']

#  transformer une cat√©gorie au format entier en plusieurs colonnes au format binaire
df_Train = pd.get_dummies(df_Train, prefix_sep='_', columns = categ)
df_test = pd.get_dummies(df_test, prefix_sep='_', columns = categ)

# convertir les colonnes (location_type,cellphone_access,gender_of_respondent) de deux dataframes au format entier
df_Train['location_type'] = lc.fit_transform(df_Train['location_type'])
df_Train['cellphone_access'] = lc.fit_transform(df_Train['cellphone_access'])
df_Train['gender_of_respondent'] = lc.fit_transform(df_Train['gender_of_respondent'])

df_test['location_type'] = lc.fit_transform(df_test['location_type'])
df_test['cellphone_access'] = lc.fit_transform(df_test['cellphone_access'])
df_test['gender_of_respondent'] = lc.fit_transform(df_test['gender_of_respondent'])
```

```
#Affichage dataframe df test avec la nouvelle transformation des donn√©es
df_test
```

```
#Affichage dataframe df train avec la nouvelle transformation des donn√©es
df_Train
```

```
#pr√©paration de x_train, Y_train et x test
X_train = df_Train.drop(['year', 'uniqueid', 'bank_account', 'country'], axis=1)
y_train = df_Train['bank_account']
X_test = df_test.drop(['year', 'uniqueid', 'country'], axis=1)

# importer MinMaxScaler (pour normaliser variables d'entr√©e,toutes les entit√©s seront transform√©es dans la plage [0,1])
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
X_train_res = scaler.fit_transform(X_train)
X_test_res = scaler.fit_transform(X_test)
```

```
df_Train
```

```
# Divis√© donn√©es de train et test
from sklearn.model_selection import train_test_split
X_Train, X_val, y_Train, y_val = train_test_split(X_train_res, y_train, stratify = y_train, test_size = 0.2, random_state=42)
```

```
#importer XGBClassifier
from xgboost import XGBClassifier
my_model = XGBClassifier()

# importer GridSearchCV
from sklearn.model_selection import GridSearchCV

# Optimiser les param√®tres du mod√®le
param_grid = {'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5]
        }
my_model2 = GridSearchCV(my_model, param_grid)
my_model2.fit(X_Train, y_Train)
print(my_model2.best_params_)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score

# √âvaluer le mod√®le
my_model3 = XGBClassifier(min_child_weight = 5, gamma = 0.5, subsample = 0.6, colsample_bytree = 0.6, max_depth = 5)
my_model3.fit(X_Train, y_Train)
y_pred = my_model3.predict(X_val)

# Obtenir le taux d'erreur
print("Error rate of Random Forest classifier: ", 1 - accuracy_score(y_val, y_pred))

# Obtenir la matrice de confusion
confusion_matrix(y_pred, y_val)
```

```
#Obtenez le r√©sultat pr√©vu pour les donn√©es de test (predicted result)
df_test.bank_account = my_model3.predict(X_test_res)
```

```
# cr√©er un DataFrame pour les nouveau donn√©es
df_result = pd.DataFrame({"uniqueid": df_test["uniqueid"] + " x " + df_test["country"],

"bank_account": df_test.bank_account})
```
